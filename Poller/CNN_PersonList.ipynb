```python
'''
    need to open name list
    
    run the titles through something to see what names are in the title
    
    output articles with names in title to folder with person's name and file cnn.txt
'''
```

```python
import os
import spacy

def returnName(line):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(line)
    
    people = []
    
    for ent in doc.ents:
        #print(ent.text, ent.label_)
        if ent.label_ == "PERSON":
            people.append(ent.text)
    
    return people


# add the title and url to the main file
def addOutputFile(titleLine, urlLine, name):    
    # makes sure there's no characters that would 
    # ruin the folder name
    if name[-1] == '\n':
        name = name.replace('\n', '')

    # get rid of the the extra CNN parts that get caught in the name finding
    if name.find("CNN"):
        name = name.replace('CNN', '')
    
    # get rid of 's in names
    if name.find("\'s"):
        name = name.replace("\'s", '')
    
    # getting rid of a list of characters that mess with the file naming structure
    char_to_replace = {'|': '',
                   ',': '',
                   ':': '',
                    '\"': '',
                    '\'': '',
                    '-': '',
                    ' ': '_'}
    
    name = name.translate(str.maketrans(char_to_replace))
    
    # creating the output file paths and names
    filePath = "ppl/" + name
    fileName = filePath + "/CNN_List.txt"
    try:
        os.mkdir(filePath)
        outputFile = open(fileName, "x")
    except:
        outputFile = open(fileName, "a")
    
    outputFile.seek(0, os.SEEK_END)
    
    toWrite = [titleLine, urlLine]
    
    outputFile.writelines(toWrite)
    outputFile.close()

    

    
    
print("\t\tStarting...")
    
inputFile = open("NameList.txt", "r")

inputFile.seek(0, os.SEEK_SET)

line = inputFile.readline()

print("\tFinding name...")
while line:

    #print(line)
    if line[:5] == "https":
        line = inputFile.readline()
        
    people = returnName(line)
    titleLine = line
    urlLine = inputFile.readline()
    
    for name in people:
        print(name)
        addOutputFile(titleLine, urlLine, name)
    
    line = inputFile.readline()

inputFile.close()

print("Done!")
```
